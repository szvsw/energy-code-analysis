```python
import shutil
from tqdm.notebook import tqdm
import os
from pathlib import Path
import pandas as pd
```

```python
data_root = Path("data")
output_root = data_root / "batches"
zip_dir = data_root / "zips"
zipped_files = os.listdir(zip_dir)


bldgs = []

bldg_idx_names = ["Country", "BuildingType", "DesignClimateZone", "HeatingType", "GroundType", "CodeVersion", "ClimateZone", "IDFPath"]
for zipfile in tqdm(zipped_files):
    if not zipfile.endswith(".zip"):
        continue
    cz = zipfile.split("CZ")[-1].split(".")[0]
    cz_dir = output_root / cz
    idf_dir = cz_dir / "idfs"
    epw_dir = cz_dir / "epws"
    os.makedirs(idf_dir, exist_ok=True)
    os.makedirs(epw_dir, exist_ok=True)
    shutil.unpack_archive(zip_dir / zipfile, idf_dir)
    for file in os.listdir(idf_dir):
        file_path = idf_dir / file
        if not file.lower().endswith(".idf"):
            os.remove(file_path)
            continue
        file_tags = file[:-4].split("+")

        file_tags.extend([cz, str(file_path) ])
        bldgs.append(file_tags)

df = pd.DataFrame(bldgs, columns=bldg_idx_names)
df["DesignID"] = range(len(df))
```
